```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#load data
tip = sns.load_dataset("tips")

tip.head()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tip)
```

```python
p1 = sns.FacetGrid(col="time", row="sex", data=tip)
p1.map(sns.scatterplot, "total_bill", "tip")
```

```python
p1 = sns.FacetGrid(col="time", row="sex", data=tip)
p1.map_dataframe(sns.histplot, x="total_bill")
```
